```python
import requests
import bs4
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://stats.espncricinfo.com/ci/content/records/93276.html"
res=requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text,"html.parser")
print(soup.prettify)
```

```python
table=soup.find("table")
print(table.text)
```

```python
rows=soup.find_all("tr",class_="data1")
print(rows)
```

```python
data=[]
for i,items in enumerate(rows):
    data.append(items.text.strip().split("\n"))
```
